```python
#load pandas
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
CO_df = pd.read_csv('Co Master Data.csv')
type (CO_df)
CO_df.corr()
CO_df.isnull().sum()
```

```python
CO_df.head()
```

```python
#correlation heatmap used in order to view correlation between CO_con and all other variables
#pleasant graph just used to get a general grasp on correlations between variables
plt.figure(figsize=(16, 6))
heatmap = sns.heatmap(CO_df.corr(), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=12)
#plt.savefig('Correlation_Heatmap.png', dpi=300, bbox_inches='tight')
```

```python
#correlation of one variable (Co_con)
#Data shows Coal consumption is the most highly correlated to CO, everything else is negative
plt.figure(figsize=(8, 12))
heatmap = sns.heatmap(CO_df.corr()[['CO_Con']].sort_values(by='CO_Con', ascending=False), 
                      vmin=-1, vmax=1, annot=True, cmap= 'BrBG')
heatmap.set_title('Variables Correlating with CO Concentration', fontdict={'fontsize':18}, pad=16);
#plt.savefig('CO_Con_Correlation_Matrix.png', dpi=300, bbox_inches='tight')
```

```python
#general graphs of certain variables with CO_con
sns.relplot(x = 'CO_Con', y = 'Tot_ Ind_ Coal', data = CO_df)
sns.relplot(x = 'CO_Con', y = 'Tot_Gas', data = CO_df)
sns.relplot(x = 'CO_Con', y = 'Net_gen_AS', data = CO_df)
sns.relplot(x = 'CO_Con', y = 'WAT', data = CO_df)
```

```python
#linegraph of CO_con over the years (requires a split of month/year columns)
sns.lineplot(x= 'Month', y = 'CO_Con', data = CO_df)
```

```python
#dist plot of CO_con
dist = sns.distplot(CO_df['CO_Con'], bins = 10)
dist.set_title ('Dist of CO_Con')
plt.savefig('Dist of CO_Con.png', dpi=300, bbox_inches='tight')
```
