```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("Covid - 19 Envir Master.csv")
pd.set_option('display.max_columns', None)
```

```python
df
```

```python
df.columns
```

```python
df_subset = df[['Month','No2_con', 'No2_AQI',
       'NG_Con', 'Tot_Petrol_AS', 'El_Util_Con_Petrol',
       'Indep_Power_Cons_Petrol', 'Tot_Com_Indus_Petrol', 'Net_gen_AS',
       'Net_gen_elec_util', 'Net_gen_Com/ind', 'Tot_Flights', 'Tot_Gas',
       'Tot_ Ind_ Coal', 'Jet_Fuel', 'GEO', 'NG', 'NUC', 'OBG', 'OT', 'SUN',
       'WAT', 'WND']]
df_subset
```

```python
sns.scatterplot(x="No2_con", y="Tot_ Ind_ Coal", data=df_subset);
```

```python
from scipy.stats import zscore
numeric_cols = df.select_dtypes(include=[np.number]).columns
df_z = df[numeric_cols].apply(zscore)
months = df["Month"]
df_z = df_z.join(months)
df_z
```

```python
df_z.plot(x="Month", y=["Tot_ Ind_ Coal", "No2_con"])
plt.ylabel('Standardized Values(Z score)', fontsize=14)
plt.title ('Standardized Changes in Coal and NO2 concentration', fontsize=16)
plt.show()
```

```python
df_z.plot(x="Month", y=["Tot_Petrol_AS", "No2_con"])
plt.ylabel('Standardized Values(Z score)', fontsize=14)
plt.title ('Standardized Changes in Petroleum and NO2 concentration', fontsize=16)
plt.show()
```

```python
df_z.plot(x="Month", y=["NG", "No2_con"])
plt.ylabel('Standardized Values(Z score)', fontsize=14)
plt.title ('Standardized Changes in Natural Gas and NO2 concentration', fontsize=16)
plt.show()
```

```python
plt.scatter(x=df_z['Tot_ Ind_ Coal'], y=df_z['No2_con'])
plt.title('Coal by NO2 Concentration Z-Scores')
plt.ylabel("NO2 Concentration(PPM)")
plt.show()
```

```python
df_z.plot(x="Month", y=["Tot_Gas", "No2_con"])
plt.ylabel('Standardized Values(Z score)', fontsize=14)
plt.title ('Standardized Changes in Total Gas and NO2 concentration', fontsize=16)
plt.show()
```

```python
plt.scatter(x=df_z['NG'], y=df_z['No2_con'])
plt.title('Natural Gas by NO2 Concentration Z-Scores')
plt.ylabel("NO2 Concentration(PPM)")
plt.show()
```

```python
plt.scatter(x=df_subset['Month'], y=df_subset['No2_AQI'])
plt.title('')
plt.xticks(rotation=90)
plt.xlabel("Months")
plt.ylabel("Air Quality Index (NO2)")
plt.show()
```

```python
sns.lmplot(x='NG',y='No2_con', data=df_z, fit_reg=True)
plt.title('Residual Plot Between No2 Concentration and Natural Gas')
```
